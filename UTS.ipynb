## Brightness
Operasi ini melakukan penambahan nilai pixel jika nilai brightness dinaikkan, dan mengurangkan nilai pixel jika nilai brightness diturunkan

g(x,y)=f(x,y)+b

◦ g(x,y) adalah nilai pixel setelah transformasi

◦ f(x,y) adalah nilai pixel asli

◦ b adalah nilai brightness

```python
# Menggunakan Linier Brightness
print('Mengubah tigkat kecerahan citra')
print('-------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')

# Program membuat variabel 'brightness_image' 
# yang ukurannya sama dengan citra original dengan tipe data yang sama dengan original.
original = cv.imread('/content/drive/MyDrive/Mechine Learning /Gambar2.jpg')
brightness_image= np.zeros(original.shape, original.dtype)

#akses per piksel
for y in range(original.shape[0]): # variabel 'y' akan merepresentasikan indeks baris
  for x in range(original.shape[1]): # variabel 'x' akan merepresentasikan indeks kolom
    for c in range(original.shape[2]): # variabel 'c' merepresentasikan indeks channel pada setiap piksel (misalnya R,G,B pada citra RGB)
      brightness_image[y,x,c] = np.clip(original[y,x,c] +  brightness, 0, 255) # Rumus dari brightness

# Program menampilkan hasilnya dengan menggabungkan citra asli 
# dan citra yang telah diubah kecerahannya menggunakan cv.hconcat().
final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

## Contrast
Kontras adalah tingkat penyebaran pixel – pixel ke dalan intensitas warna

g(x,y)=a* f(x,y)+b

◦ g(x,y) adalah nilai pixel setelah transformasi

◦ f(x,y) adalah nilai pixel asli

◦ a adalah nilai contrast

◦ b adalah nilai brightness

```python
# Menggunakan kontras
print('Mengubah kontras citra')
print('-------------------------------')
try:
  brightness = int(input('Masukkan tingkat kecerahan [-255 - 255]: '))
  kontras = int(input('Masukkan kontras [1.0 - 3.0]: '))
except ValueError:
  print('Error, not a number')

# Program membuat variabel 'inverse_image' 
# yang ukurannya sama dengan citra original dengan tipe data yang sama dengan original.
original = cv.imread('/content/drive/MyDrive/Mechine Learning /Gambar2.jpg')
inverse_image= np.zeros(original.shape, original.dtype)

#akses per piksel
for y in range(original.shape[0]):# variabel 'y' akan merepresentasikan indeks baris
  for x in range(original.shape[1]):# variabel 'x' akan merepresentasikan indeks kolom
    for c in range(original.shape[2]):# variabel 'c' merepresentasikan indeks channel pada setiap piksel (misalnya R,G,B pada citra RGB)
      inverse_image[y,x,c] = np.clip(kontras * original[y,x,c] +  brightness, 0, 255) # rumus dari kontras

# Program menampilkan hasilnya dengan menggabungkan citra asli 
# dan citra yang telah diubah kecerahannya menggunakan cv.hconcat().
final_frame = cv.hconcat((original, inverse_image))
cv2_imshow(final_frame)
```

## Logarithmic Brightness
Transformasi Log memetakan suatu citra dengan range warna sempit menjadi lebih lebar pada citra output

s=c* log(1+r)

◦ c : konstanta

◦ r : nilai grey-level citra input

◦ s : nilai grey-level citra output

```python
# Logarithmic Brightness
print('Mengubah tigkat kecerahan citra dengan transformasi log')
print('-------------------------------------------------------')
try:
  logarithmic_brightness = int(input('Masukkan nilai kontanta: '))
except ValueError:
  print('Error, not a number')

# Program membuat variabel 'logarithmicBrightness_image' 
# yang ukurannya sama dengan citra original dengan tipe data yang sama dengan original.
original = cv.imread('/content/drive/MyDrive/Mechine Learning /Gambar2.jpg')
logarithmicBrightness_image= np.zeros(original.shape, original.dtype)

#akses per piksel
for y in range(original.shape[0]): # variabel 'y' akan merepresentasikan indeks baris
  for x in range(original.shape[1]): # variabel 'x' akan merepresentasikan indeks kolom
    for c in range(original.shape[2]): # variabel 'c' merepresentasikan indeks channel pada setiap piksel (misalnya R,G,B pada citra RGB)
      logarithmicBrightness_image[y,x,c] = np.clip(logarithmic_brightness * (np.log(1 + original[y,x,c])), 0 , 255) # Rumus dari logarimict brightness

# Program menampilkan hasilnya dengan menggabungkan citra asli 
# dan citra yang telah diubah kecerahannya menggunakan cv.hconcat().
final_frame = cv.hconcat((original, logarithmicBrightness_image))
cv2_imshow(final_frame)
```

## Inverse Color
Proses yang digunakan untuk membalik nilai citra dengan cermin nilai pixel tengah

g(x)=255-f(x)

◦ g(x,y) adalah citra negative

◦ f(x,y) adalah citra asli

```python
print('Mengubah inverse citra')
print('-------------------------------')
try:
  inverse = int(input('Masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')

# Program membuat variabel 'inverse_image' 
# yang ukurannya sama dengan citra original dengan tipe data yang sama dengan original.
original = cv.imread('/content/drive/MyDrive/Mechine Learning /Gambar2.jpg')
inverse_image= np.zeros(original.shape, original.dtype)

#akses per piksel
for y in range(original.shape[0]): # variabel 'y' akan merepresentasikan indeks baris
  for x in range(original.shape[1]): # variabel 'x' akan merepresentasikan indeks kolom
    for c in range(original.shape[2]): # variabel 'c' merepresentasikan indeks channel pada setiap piksel (misalnya R,G,B pada citra RGB)
      inverse_image[y,x,c] = np.clip(255 - original[y,x,c], 0, 255) # Rumus dari inverse

# Program menampilkan hasilnya dengan menggabungkan citra asli 
# dan citra yang telah diubah kecerahannya menggunakan cv.hconcat().
final_frame = cv.hconcat((original, inverse_image))
cv2_imshow(final_frame)
```

## Grayscale
Grayscale adalah citra keabuan dengan nilai intensitas kedalaman pixel 8 bit

1. Grayscale Average = (R + G + B)/3
2. Grayscale Lightness = Max[R,G,B] + Min[R,G,B]/2
3. Grayscale Luminance = 0.21R + 0.72G + 0.07B 

```python
# MENAMPILKAN GAMBAR
img = cv.imread('/content/drive/MyDrive/Mechine Learning /Gambar2.jpg')
img_rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB) # konversi channel BGR -> RGB
plt.imshow(img_rgb)
```

```python
# Metode Averaging
img_Averaging = cv.imread('/content/drive/MyDrive/Mechine Learning /Gambar2.jpg')

# menggunakan fungsi mean untuk mengitung rata-rata
# argumen axis=-1 dan keepdims=1 digunakan untuk menjaga dimensi array, 
#sehingga menghasilkan array 2 dimensi dengan setiap elemen bernilai 
# rata-rata kecerahan dari tiga channel warna RGB.
img_Averaging[:] = img_Averaging.mean(axis=-1, keepdims=1)

# membuat sebuah objek plot (figure) yang memiliki dua area plot (subplots), 
# dengan ukuran 1x2 (satu baris, dua kolom). 
fig, ax = plt.subplots(1,2)

# Gambar asli
ax[0].imshow(img_rgb)

# Gambar setelah diedit
ax[1].imshow(img_Averaging)
```

```python
# Metode lightness
img_lightness = cv.imread('/content/drive/MyDrive/Mechine Learning /Gambar2.jpg')

# menggunakan fungsi max dan min
img_lightness[:] = np.max(img_lightness,axis=-1, keepdims=100)/2 + np.min(img_lightness,axis=-1, keepdims=100)/2

# membuat sebuah objek plot (figure) yang memiliki dua area plot (subplots), 
# dengan ukuran 1x2 (satu baris, dua kolom). 
fig, ax = plt.subplots(1,2)

# Gambar asli
ax[0].imshow(img_rgb)

# Gambar setelah diedit
ax[1].imshow(img_lightness)
```

```python
# Metode luminance
img_luminance = cv.imread('/content/drive/MyDrive/Mechine Learning /Gambar2.jpg')

# mengakses komponen RGB pada citra menggunakan slicing [:,:,0], [:,:,1], [:,:,2] 
# pada variabel img_luminance dan menyimpannya pada masing-masing variabel R, G, B.
R, G, B = img_luminance[:,:,0], img_luminance[:,:,1], img_luminance[:,:,2]
gray_img_luminance = (0.21 * R) + (0.72 * G) + (0.07 * B)

# membuat sebuah objek plot (figure) yang memiliki dua area plot (subplots), 
# dengan ukuran 1x2 (satu baris, dua kolom). 
fig, ax = plt.subplots(1,2)

# Gambar asli
ax[0].imshow(img_rgb)

# Gambar setelah di edit 
# Pada argumen cmap='gray' digunakan untuk menampilkan 
#citra grayscale dengan warna abu-abu (gray) pada objek plot.
ax[1].imshow(gray_img_luminance, cmap='gray')
```
